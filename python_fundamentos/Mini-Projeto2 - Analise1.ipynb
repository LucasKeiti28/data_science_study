## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('dark_background')
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Distribuição de Veículos com base no Ano de Registro

```python
df.head()
df.info()
df.count()
df.describe()
df.columns[:27]
```

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
fig, ax = plt.subplots(figsize=(8,6))
sns.distplot(df['yearOfRegistration'], color="#7159c1", kde=True, ax = ax)

plt.title('Distribuicao de Veiculos com base no ano de registro.')
plt.ylabel('Densidade de Veiculos', fontsize=15)
plt.xlabel('Ano de Registro', fontsize=15)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
# sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(8,6))
sns.boxplot(x='vehicleType', y='price', data=df)
plt.title('Analise de Outliers')
plt.xlabel('Tipo de Veiculo')
plt.ylabel('Preco')
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/price-vehicleType-boxplot.png")
```
